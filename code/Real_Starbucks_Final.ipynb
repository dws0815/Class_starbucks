```python
menu_list = {'아메리카노':4500, '라떼':5000, '카라멜 마끼아또':6000}
bean_list = {'아메리카노':50, '라떼':50, '카라멜 마끼아또':50}
milk_list = {'아메리카노':0, '라떼':400, '카라멜 마끼아또':400}
syrup_list = {'아메리카노':0, '라떼':50, '카라멜 마끼아또':100}
cup_list = {'아메리카노':1, '라떼':1, '카라멜 마끼아또':1}

## 날짜별로 분류하기 위한 모듈 (+ timedelta와 랜덤 함수를 통해 날짜를 차감함, orders에 timedelta 확인)

from datetime import datetime,date,timedelta 

class Starbucks:

    def __init__(self, name):
        self.name = name
        self.sales = 0
        self.cnt = 0
        self.bean = 0
        self.milk = 0
        self.syrup = 0
        self.cup = 0
        self.menues=[]
        self.times=[]
        self.quantities=[]
        self.prices=[]
        print('스타벅스 지점명:', name,'\n')
        
    def order(self, menu, quantity):
        chk=0
        for keys in menu_list:
            if menu == keys:
                price = int(menu_list[keys])
                self.sales += price*quantity
                
                self.menues.append(menu)
                self.times.append((datetime.today()).strftime("%Y/%m/%d"))
                self.quantities.append(quantity)
                self.prices.append(price*quantity)
                
                bean = int(bean_list[keys])
                self.bean += bean*quantity
                
                
                milk = int(milk_list[keys])
                self.milk += milk*quantity
                
                
                syrup = int(syrup_list[keys])
                self.syrup += syrup*quantity
                
                
                cup = int(cup_list[keys])
                self.cup += cup*quantity
        
        
                self.cnt += 1
                print(datetime.today().strftime("%Y/%m/%d %H:%M:%S"),"%s 주문번호: %s, %s %s잔 주문완료 \t" % ( self.name, self.cnt, menu, quantity))
                chk += 1
                print()
        
    
    def orders(self, menu, quantity,i):
        chk=0
        for keys in menu_list:
            if menu == keys:
                price = int(menu_list[keys])
                self.sales += price*quantity
                
                self.menues.append(menu)
                self.times.append((datetime.today() - timedelta(i)).strftime("%Y/%m/%d"))
                self.quantities.append(quantity)
                self.prices.append(price*quantity)
                
                bean = int(bean_list[keys])
                self.bean += bean*quantity
                
                
                milk = int(milk_list[keys])
                self.milk += milk*quantity
                
                
                syrup = int(syrup_list[keys])
                self.syrup += syrup*quantity
                
                
                cup = int(cup_list[keys])
                self.cup += cup*quantity
        
        
                self.cnt += 1
                print(datetime.today().strftime("%Y/%m/%d %H:%M:%S"),"%s 주문번호: %s, %s %s잔 주문완료 \t" % ( self.name, self.cnt, menu, quantity))
                chk += 1
                print()
                
                
        if chk != 1:
                print(menu, "는 없는 메뉴입니다")
                print()

        
    def show_sales(self):
        print("*" * 60) 
        print("현재시각:", datetime.today().strftime("%Y/%m/%d %H:%M:%S"))
        print('스타벅스 %s의 현재까지 주문 수: %s, 매출: %d원' %( self.name, self.cnt, self.sales)) 
        
    
    def show_stock_consumption(self):
        print("*" * 60) 
        print("현재시각:", datetime.today().strftime("%Y/%m/%d %H:%M:%S"))
        print('스타벅스 %s의 현재 발생한 누적재고' %(self.name))
        print('원두: %sg' %(self.bean) )
        print('우유: %sml' %(self.milk) )
        print('시럽: %sg' %(self.syrup) )
        print('컵: %s개' %(self.cup) )
        
        
    def show_stock_status(self, bean, milk, syrup, cup):
        print("*" * 60) 
        print("현재시각:", datetime.today().strftime("%Y/%m/%d %H:%M:%S"))
        print('스타벅스 %s의 재고 현황' %(self.name))
        
        if self.bean < bean:
            print('원두: %sg' %( bean - self.bean) )
        else:
            print("원두 재고가 부족합니다(발주 필요)")
        if self.milk < milk:
            print('우유: %sml' %( milk - self.milk) )
        else:
            print("시럽 재고가 부족합니다(발주 필요)")
        if self.syrup < syrup:
            print('시럽: %sg' %( bean - self.bean) )
        else:
            print("시럽 재고가 부족합니다(발주 필요)")
        if self.cup < cup:
            print('컵: %s개' %( cup - self.cup) )
        else:
            print("컵 재고가 부족합니다(발주 필요)")
            
def show_sales_difference(매장1, 매장2):
    비교=매장1.sales-매장2.sales
    if 매장1.sales>매장2.sales:
        print("*" * 60)
        print("두 지점 중 매출액이 큰 매장은 ", 매장1.name, " 입니다.\n차액은 ", 매장1.sales-매장2.sales," 원 입니다.")
    elif 매장1.sales==매장2.sales:
        print("*" * 60)
        print("두 지점의 매출액은 같습니다.")
    else:
        print("*" * 60)
        print("두 지점 중 매출액이 큰 매장은 ", 매장2.name ," 입니다.\n차액은 ", 매장2.sales-매장1.sales," 원 입니다.")
        
```

```python
### 클래스 할당하기
가산현대아울렛점=Starbucks('가산현대아울렛점')
마리오아울렛점=Starbucks('마리오아울렛점')
```
